```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

```python
# -------------------------
# Data
# -------------------------
labels = ["MeanDiff", "PCA k=8", "PCA k=32", "PCA k=64", "PCA k=128"]

orig_loss = np.array([11.0084, 11.0084, 11.0084, 11.0084, 11.0084])
patched_loss = np.array([11.1086, 10.6128, 10.2912, 9.9908, 9.5268])

# Loss SE only provided for MeanDiff table.
# Replace PCA zeros if you have PCA loss SEs.
orig_loss_se = np.array([0.0055, 0, 0, 0, 0])
patched_loss_se = np.array([0.0054, 0, 0, 0, 0])

orig_rank = np.array([624.7, 624.7, 624.7, 624.7, 624.7])
patched_rank = np.array([1216.7, 2073.1, 3818.9, 3550.5, 3558.8])

# Rank SE not provided for MeanDiff patched row, so set to 0.
orig_rank_se = np.array([0, 3.4216, 3.4216, 3.4216, 3.4216])
patched_rank_se = np.array([0, 8.7705, 12.5685, 11.8902, 12.1811])

# Colors
original_color = "lightgray"
meandiff_color = "tab:blue"
pca_colors = ["#fcae91", "#fb6a4a", "#de2d26", "#a50f15"]

patched_colors = [meandiff_color] + pca_colors

x = np.arange(len(labels))
width = 0.36
```

```python


# -------------------------
# Chart 1: Loss
# -------------------------
plt.figure(figsize=(8, 5))

plt.bar(
    x - width / 2,
    orig_loss,
    width,
    yerr=orig_loss_se,
    capsize=4,
    label="Original",
    color=original_color,
    edgecolor="black",
)

plt.bar(
    x + width / 2,
    patched_loss,
    width,
    yerr=patched_loss_se,
    capsize=4,
    label="Patched",
    color=patched_colors,
    edgecolor="black",
)

plt.ylabel("Loss")
plt.xlabel("Method")
plt.title("Original vs Patched Loss")
plt.xticks(x, labels, rotation=0)
plt.tight_layout()
plt.savefig("loss_bar_chart.png", dpi=300)
plt.show()

```

```python
# -------------------------
# Chart 2: Mean Rank
# -------------------------
plt.figure(figsize=(8, 4))

plt.bar(
    x - width / 2,
    orig_rank,
    width,
    yerr=orig_rank_se,
    capsize=4,
    label="Original",
    color=original_color,
    edgecolor="black",
)

plt.bar(
    x + width / 2,
    patched_rank,
    width,
    yerr=patched_rank_se,
    capsize=4,
    label="Patched",
    color=patched_colors,
    edgecolor="black",
)

plt.ylabel("Mean Rank")
plt.xlabel("Method")
plt.title("Original vs Patched Mean Rank")
plt.xticks(x, labels, rotation=0)
plt.tight_layout()
plt.savefig("mean_rank_bar_chart.png", dpi=300)
plt.show()
```
